# Load dataset

```python
df_movies = pd.read_csv('dataset/raw/tmdb_5000_movies.csv')
df_credits = pd.read_csv('dataset/raw/tmdb_5000_credits.csv')
```

```python
df_credits = df_credits.rename(columns={'movie_id': 'id'})

# Merge dataframes
df_data = pd.merge(df_movies, df_credits, on='id')
df_data.head()
```

### Drop some attributes

Since the __homepage__ column has a large of empty values and doesn't has big importance on recommendation system, it can be dropped. The __tagline__ column also can be dropped because the most part of the movies has an overview and thus the tagline would result in redundant data. The __id__ column is unique for each movie, it is only for purpose of identification, thus it will be dropped because it not contribute to the recommendations.

```python
df_data = df_data.drop(columns=['homepage', 'tagline', 'id', 'title_x', 'title_y'])
```

```python
df_data.isnull().sum()
```

### Drop movies which are unreleased

```python
df_data['status'].value_counts()
```

```python
df_data = df_data.drop(df_data[df_data.status != 'Released'].index)
df_data = df_data.reset_index(drop=True)
```

### Converts some attributes

Convert the attributes __genres__, __keywords__, __cast__ and __crew__ to list of dictionarys.

```python
columns = ['genres', 'keywords', 'cast', 'crew']
for column in columns:
    df_data[column] = df_data[column].apply(literal_eval)
```

### Functions

Some functions to extract and manipulate the data.

```python
def get_director_name(list_data):
    for data in list_data:
        if data['job'] == 'Director':
            return data['name']
    return np.nan

def get_list(list_data):
    if isinstance(list_data, list):
        list_names = [data['name'] for data in list_data]
        return list_names
    return []

def standardize_data(obj):
    if isinstance(obj, list):
        return [str.lower(i.replace(' ', '')) for i in obj]
    else:
        if isinstance(obj, str):
            return str.lower(obj.replace(' ', ''))
        else:
            return ''
        
def create_metadata(obj):
    return ' '.join(obj['genres']) + ' ' + ' '.join(obj['keywords']) + ' ' + ' '.join(obj['cast']) + ' ' + obj['director']
        
```

```python
# Make a new column with director names.
df_data['director'] = df_data['crew'].apply(get_director_name)
```

```python
# Convert the columns below to list of names.
columns = ['genres', 'keywords', 'cast']
for column in columns:
    df_data[column] = df_data[column].apply(get_list)
```

```python
# Convert all strings to lowercase and remove spaces.
columns = ['genres', 'keywords', 'cast', 'director']
for column in columns:
    df_data[column] = df_data[column].apply(standardize_data)
```

### Generate the metadata

The columns __genres__, __keywords__, __cast__ and __director__ are used to generate the metadata. These metadata will be used to feed the word vector model.


```python
# Generate the metadata that will be fed into the word vector model.
df_data['metadata'] = df_data.apply(create_metadata, axis=1)
```
